```python
import pandas as pd
import numpy as np

class Multi_log_reg:
    def softmax(self, X):
        X = X - np.max(X)
        return np.exp(X) / (np.sum(np.exp(X)))
    
    def  multi_logistic_train(self, digits, labels, num = 10, learning_rate = 0.1, batchSize = 20):
        self.weights = np.random.uniform(0, 1, (10, 28**2))
        for iter in range(num):
            batch_index =  np.random.choice(len(digits), batchSize, replace=False)
            for i in batch_index:
                x = digits[i].reshape(-1, 1)
                y = np.zeros((10, 1))
                y[labels[i]] = 1
                y_tilt = self.softmax(self.weights@x)
                self.weights -= learning_rate * ((y_tilt - y)@ x.T)
                
        return self.weights

    def multi_logistic_predict(self, digit):   
        return np.argmax(self.weights@ digit)  
```
